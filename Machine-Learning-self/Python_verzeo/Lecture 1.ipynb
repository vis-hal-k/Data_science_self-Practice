### 4. Introduction to Python

High Level - A low-level language is a language that is close to the machine. Hence, it has quite the steep learning curve. You know the machine understands zeros and ones. So imagine having to write lines of code that is almost in that binary form. You need to check the memory usage of your code, how your code is stored, how that affects performance, write some algorithms to improve performance of the program. In Python, no need to take care about low-level details such as managing the memory used by the program. Higher-level means it’s more readable to humans and less readable to computers. Likewise, Lower-level means less readable for humans and more readable for computers.

Interpreted Language - Interpreted in simple terms means running code line by line. It also means that the instruction is executed without earlier compiling the whole program into machine language. There are no separate compilation and execution steps like C and C++. Internally, Python converts the source code into an intermediate form called bytecodes which is then translated into native language of specific computer to run it. If you commit a mistake at the end of your program, you will find that there is an error generated for obvious reasons but along with the error, you will find the output of the program till the line of the program was correct.

Dynamically Typed - Everything in Python is an Object. No need to declare anything. An assignment statement binds a name to an object, and the object can be of any type.

Object-oriented language - As mentioned, everything in python is an object and you can find the datatype of an object using type()

### 5. Comments and Statements

Comments are lines that exist in computer programs that are ignored by compilers and interpreters.

Including comments in programs makes code more readable for humans as it provides some information or explanation about what each part of a program is doing.

In general, it is a good idea to write comments while you are writing or updating a program as it is easy to forget your thought process later on, and comments written later may be less useful in the long term.

In Python, we use the hash (#) symbol to start writing a comment.

```python
#Print (Hello, world) to console
print("Hello, world 2") #this to print Hello world
```

If we have comments that extend multiple lines, one way of doing it is to use hash (#) in the beginning of each line.

Docstring is short for documentation string.

It is a string that occurs as the first statement in a module, function, class, or method definition.

```python
def double(num):
    """
    function to double the number
    """
    return 2 * num

print(double(10))
```

```python
print(double.__doc__) #Docstring is available to us as the attribute __doc__ of the function
```

### 8. Variables

A variable is a location in memory used to store some data (value).

They are given unique names to differentiate between different memory locations. The rules for writing a variable name is same as the rules for writing identifiers in Python.

We don't need to declare a variable before using it. In Python, we simply assign a value to a variable and it will exist. We don't even have to declare the type of the variable. This is handled internally according to the type of value we assign to the variable.

```python
#We use the assignment operator (=) to assign values to a variable

a = 10
b = 5.5
c = "ML"
```

```python
a, b, c = 10, 5.5, "ML"
a = b = c = "AI" #assign the same value to multiple variables at once
```

```python
x = 3

print(id(x))               #print address of variable x
```

```python
y = 3

print(id(y))           #print address of variable y
```

Observation:

x and y points to same memory location

```python
y = 2
print(id(y))               #print address of variable y
```

So, the ID is corresponding to the value and not the identifier or variable

```python
p = 25
print(id(p))
```

```python
p = 257
q = 257
print(id(p))
print(id(q))
```

```python
q = 25
print(id(q))
```

```python
p = 257
q = 257

print(id(p), id(q))
```

```python
a = 10; b = 20  #multiple statements in single line

print("The value of b is {1} and a is {0}".format(a, b)) #specify position of arguments
```
