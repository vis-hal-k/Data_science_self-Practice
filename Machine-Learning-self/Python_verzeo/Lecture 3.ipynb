### Control Flow While Loop

The while loop in Python is used to iterate over a block of code as long as the test expression (condition) is true.

Syntax:

while test_expression:

    Body of while

The body of the loop is entered only if the test_expression evaluates to True.

After one iteration, the test expression is checked again.

This process continues until the test_expression evaluates to False.

```python
#Find product of all numbers present in a list

lst = [10, 20, 30, 40, 60]

product = 1
index = 0

while index < len(lst):
    product *= lst[index] # p=p*lst[index]
    index +=1 # most imp

print("Product is: {}".format(product))
```

### while loop with else

we can have an optional else block with while loop as well.

The else part is executed if the condition in the while loop evaluates to False. 

The while loop can be terminated with a break statement. In such case, the else part is ignored. Hence, a while loop's else part runs if no break occurs and the condition is false.

```python
numbers = [1, 2, 3,4,5]

#iterating over the list
index = 0
while index < len(numbers):
    print(numbers[index])
    index += 1
    
else:
    print("no item left in the list")
```

```python
# But if I break the above loop before test condition evaluates to False

numbers = [1, 2, 3,4,5]

index = 0
while index < len(numbers):
    print(numbers[index])
    index += 1
    if numbers[index] == 3:
        break
    
else:
    print("no item left in the list")
```

### Range Function

We can generate a sequence of numbers using range() function. range(10) will generate numbers from 0 to 9 (10 numbers).

We can also define the start, stop and step size as range(start,stop,step size). step size defaults to 1 if not provided.

This function does not store all the values in memory, it would be inefficient. So it remembers the start, stop, step size and generates the next number on the go. Range does not generates all numbers that it contains when we create it. It gives only those numbers which we get them using loop.

Range is not a generator. But the behaviour of Generator is also the same. It gives us elements one by one.

What is a Generator?

Generators are basically functions that return traversable objects or items. These functions do not produce all the items at once, rather they produce them one at a time and only when required. Whenever the for statement is included to iterate over a set of items, a generator function is run. Generators have a number of advantages as well.

Memory is saved as the items are produced when required, unlike normal Python functions. This fact becomes very important when you need to create a huge number of iterators. This is also considered as the biggest advantage of generators. At a time, only one element is actually there in the RAM.

Generator functions allow you to declare a function that behaves like an iterator. They allow programmers to make an iterator in a fast, easy, and clean way. What’s an iterator, you may ask? An iterator is an object that can be iterated (looped) upon. It is used to abstract a container of data to make it behave like an iterable object. You probably already use a few iterable objects every day: strings, lists, and dictionaries to name a few.

Generators have a yield method instead of return which is used in a simple function. Generators are also defined like simple Python functions only.

The difference is that while a return statement terminates a function entirely, yield statement pauses the function saving all its states and later continues from there on successive calls.

```python
def check_prime(number):  
    if (number == 1) or (number == 2):
        return False
    for divisor in range(2, number):        #int(number ** 0.5) + 1
        if number % divisor == 0:            
            return False  
    return True
```

```python
for i in range(10):
    print(i)
```

```python
def Primes(maxm):    
    number = 1    
    while number < maxm:        
        number += 1        
        if check_prime(number):            
            yield number
```

### for loop with else

A for loop can have an optional else block as well. The else part is executed if the items in the sequence used in for loop exhausts.

break statement can be used to stop a for loop. In such case, the else part is ignored.

Hence, a for loop's else part runs if no break occurs and the loop has gone over all the elements

```python
lst = ["satish", "srinu", "murali", "naveen", "bramha"]

for ele in lst:
    print(ele)
    if ele == "naveen":
        break
        
else:
    print("Elements finished!")
```

### Python Program to display all prime numbers within an interval

```python
index1 = 3
index2 = 50

print("Prime numbers between {0} and {1} are :".format(index1, index2))

for num in range(index1, index2+1):      #default step size is 1
    if num > 2:
        isDivisible = False
        for index in range(2, num):
            if num % index == 0:
                isDivisible = True
        if not isDivisible:        
            print(num)
    
```

### List Comprehension

Python List comprehension provides a much more short syntax for creating a new list based on the values of an existing list.

Advantages of List Comprehension
1. More time-efficient and space-efficient than loops.
2. Require fewer lines of code.
3. Transforms iterative statement into a formula.

Syntax of  List Comprehension

newList = [ expression(element) for element in oldList if condition ] 

```python
a = list(range(20))
```

```python
a
```

```python
even_list =  []
odd_list = []
for ele in a:
    if ele%2 == 0:
        even_list.append(ele)
    else:
        odd_list.append(ele)
```

```python
print(even_list)
print(odd_list)
```

```python
# list of even numbers
lis = [i for i in a if i % 2 == 0]
print(lis)
```

```python
odd_lis = [i for i in a if i % 2 != 0]
print(odd_lis)
```

```python
# If Else in List Comprehension

lis = [i if i % 2 == 0 else -1 for i in a]
print(lis)
```

```python
matrix = []

for i in range(2):

 # Append an empty sublist inside the list
    matrix.append([])
    if i==1:
        for j in range(5):
            matrix[i].append(j)
    else:
        for j in range(3):
            matrix[i].append(j)
        

print(matrix)
```

```python
# Nested list comprehension
matrix = [[j for j in range(4)] for i in range(3)]

print(matrix)
```
