## Pyplot
Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
Xp = np.array([0,6])
Yp = np.array([0,250])
plt.plot(Xp , Yp)
plt.show()
# Example
# Draw a line in a diagram from position (1, 3) to position (8, 10)
X = np.array([1,8])
Y = np.array([3,10])
plt.plot(X,Y, 'o')
plt.show()
```

```python
# Example
# Draw a line in a diagram from position (1, 3) to position (8, 10)
X = np.array([1,8])
Y = np.array([3,10])
plt.plot(X,Y ,':')
plt.show()
```

## Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y-points.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints,  'o:b') # o:b means o ->  circle(marker) , : -> Dotted (Line) , b -> blue(colour) . 
plt.show()
'''
for y=3,8,1,10,5,7
    x=1,2,3,4 ,5,7
'''
```

##  Marker Size
You can use the keyword argument markersize or the shorter version, "ms" to set the size of the markers:
## Marker Color
You can use the keyword argument markeredgecolor or the shorter "mec" to set the color of the edge of the markers:
#### You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the marker

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y , marker='o',ms =20 , mec= 'r' , mfc ='b')
plt.show()
```

## Multiple Line
ou can plot as many lines as you like by simply adding more plt.plot() functions:

### Example
Draw two lines by specifying a plt.plot() function for each line:

```python
import matplotlib.pyplot as plt
import numpy as np
x1= np.array([1,3,8 , 10])
y1 = np.array([3, 8, 8, 10])
x2 = np.array([2,5,7,10])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1,y1 , label="x1,y1")
plt.plot(x2,y2, label="x2,y2")
plt.legend() # it shows the label , legend parameter is location, shadow:true , fontsize etc
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y , marker= 'o' , c = 'r' ,mec = 'b' ) 
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("sports watch Data")
plt.show()
```
