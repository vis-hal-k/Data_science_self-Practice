    <h1 style = "background-color: Black; color:white; display:flex; padding:10px">Reshaping of Numpy Array</h1>
 </div>
 <p> Reshaping means changing the shape of an array. The shape of an array is the number of elements in each dimension. By reshaping we can add or remove dimensions or change number of elements in each dimension.</p>
 <p style = "font-family:Cursive;"> For example</p>


````python
import numpy as np
arr = np.random.randint(1,100,12)
# 1 to 100 ; 12 integer
arr1=arr.reshape(4,3) # convert 1d to 2d; 4x3,6x2,12x1 = 12 , similary
print(arr) # here arr is not changing/reshape in originally
print(arr1)
#convert 2d to 1d
arr2 = arr.reshape(12)
print(arr2)
````

````python
arr3= arr.reshape(3,-1) # here -1 means its automatically understand the possible value i.e 4 with 3 , (3x4=12) 
arr3
````

    <h1 style = "background-color: Black; color:white; display:flex; padding:10px">SEED function of Numpy Array</h1>
 </div>
 <p>
    we know that randint() function generates random numbers. Everytime we run the program, new set of random number is generated. But what if we want to fix the generation of this random number. In this case we use seed, this generates fix set of random numbers. 
</p>

``` Pyhton
    import numpy as np
    np.random.seed(value)
    arr=np.random.randint(1,100,10)
```
Now it will return the same random random numbers till we change the integer as passed to seed at any computer. 
<p style = "font-family:Cursive;"> For example</p>

````python
import numpy as np
np.random.seed(2)
arr=np.random.randint(1,100,10)
arr # here value is same for all syste
````

## Creating 2 Dimensional List / Array
by using for loops (manaully)

 Making 2d __List__


Making 2d __array__

````python
import numpy as np
matrix = []
row=3 ; col = 4
for i in range(row):
    a=[]
    for j in range(col):
        val=int(input())
        a.append(val)
    matrix.append(a)
arr = np.array(matrix)
for i in range(row):
    for j in range(col):
        print(arr[i][j],end=" ")
    print()    
````
