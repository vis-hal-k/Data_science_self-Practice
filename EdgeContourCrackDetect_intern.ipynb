## Canny Edge Detection 

```python
import cv2
from google.colab.patches import  cv2_imshow 
from  matplotlib import pyplot as plt 
import numpy as np 
img2 = "/content/drive/MyDrive/Colab Notebooks/Open_cv_Course/00001.jpg"
i_neu = cv2.imread(img2 , 0 )
# i_neu = cv2.resize(i_neu , None , fx = 0.5 , fy = 0.5 , interpolation = cv2.INTER_AREA)
# cv2_imshow(i_neu)
# Canny Edge Detection 
edges = cv2.Canny(i_neu, 100 , 200 )
#  100 -> minimum  value of canny
#  200 -> maximum value of canny
cv2_imshow(np.hstack((i_neu , edges)))
```

```python
import cv2 
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt 
import numpy as np 
img2 = "/content/drive/MyDrive/Colab Notebooks/Open_cv_Course/00001.jpg"
i_samp =  cv2.imread(img2)
i_samp = cv2.resize(i_samp , None , fx = 0.5 , fy = 0.5 , interpolation = cv2.INTER_AREA)
original1 = i_samp.copy()
i_samp = cv2.bilateralFilter(i_samp , 7,80 , 100)
cv2_imshow(i_samp)
ig_samp = cv2.cvtColor(i_samp , cv2.COLOR_BGR2GRAY)
# _,thresh=cv2.threshold(ig_samp , 127,225, cv2.THRESH_BINARY)
edges = cv2.Canny(ig_samp , 100 , 200 )
# contours, hierarchy=cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)  
contours, hierarchy=cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)  
cv2.drawContours(i_samp, contours ,-1 , (0,255,0 ) ,  1 )
cv2_imshow(edges)
# cv2_imshow(thresh) 
cv2_imshow(i_samp)
cv2_imshow(np.hstack((original1 , i_samp)))
```
