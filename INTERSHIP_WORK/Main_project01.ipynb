```python
Cracks = "/content/drive/MyDrive/Colab Notebooks/Internship/Cracks_largePic.jpg"
```

```python
import patchify
from PIL import Image 
from matplotlib import pyplot as plt 
import numpy as np 
```

```python
i_crack = Image.open(Cracks)
print(f"Size =  {i_crack.size} \nmode = {i_crack.mode} \nheight = {i_crack.height} \nwidth = {i_crack.width}")
```

```python
ip_crack = i_crack.resize((3024,4032))
# ip_crack = i_crack.resize((2805,3825))
```

```python
gather=[]
patches = patchify.patchify(np.asarray(ip_crack) , patch_size=(255,255,3) , step = 255)
patches.shape
```

```python
for i in range (patches.shape[0]):
  for j in range (patches.shape[1]):
    get_patches = patches[i,j,0,:,:,:]
    gather.append(get_patches)
display(len(gather)) # 11*15
```

```python
fig1, ax1 = plt.subplots(nrows = 15 , ncols = 11 , figsize=(18,16))
a=0 
for R1 in range(15):
  for C1 in range(11):
    ax1[R1,C1].imshow(gather[a])
    ax1[R1,C1].axis("off")
    a+=1 
```
