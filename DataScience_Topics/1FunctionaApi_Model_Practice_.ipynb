```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import *
```

```python
from keras.models import Model
```

## Simple Basic Level Examples -> Functinal Api 

```python
model =Model(inputs = x, outputs = [output1 , output2])
```

```python
# input is tabular of 3 columns
x = keras.Input(shape=(3,))
#  Now we create Hidden layer1. Dense(  , activation , )(x) ; -> (x) is just connection to input layer.  
hidden1 = Dense(128,activation='relu')(x)            #Dense
hidden2 = Dense(64,activation = 'relu')(hidden1)     # Dense1
# Now to do Branching
output1 = Dense(1,activation='linear')(hidden2)      # Dense 2

output2 = Dense(1,activation = 'sigmoid')(hidden2)   # Dense 3
# this out model architecture.

```

```python
model.summary()
```

```python
# To  visualize what we create , their is utlis in keras .
# from keras.utils import plot_model
# plot_model(model)
tf.keras.utils.plot_model(
    model,
    # tofile 
    show_shapes = True,
    show_dtype = True ,
    # show_layer_names = True
    rankdir = "TB",
    expand_nested =False,
    dpi = 80, # It for size , dpi -> dots per inch 
    # layer_range = None 
    show_layer_activations = False ,
    show_trainable = True
)
```

```python
inputA = keras.Input(shape=(32,))
inputB = keras.Input(shape=(128,))
```

```python
#  The first branch operates on the first input
b1a = Dense(8, activation='relu')(inputA )
b1b = Dense(4, activation='relu')(b1a)
```

```python
b2a = Dense(64 , activation ="relu")(inputB) 
b2b = Dense(32 , activation = "relu")(b2a)
b2c = Dense(4, activation = "relu")(b2b)
# combine the output of the twp branches.. 
b3con  = concatenate([b1b,b2c])
b3a = Dense(8 , activation ='relu')(b3con)
b3b = Dense(2,activation = "relu")(b3a)

```

```python
model2 = Model(inputs = [inputA , inputB], outputs = b3b)
```

```python
tf.keras.utils.plot_model(
    model2 ,
    show_shapes = True,
    show_dtype = True ,
    show_layer_activations = True,
    dpi = 89
)
```
