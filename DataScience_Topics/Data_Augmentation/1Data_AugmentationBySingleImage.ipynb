# Data Augmentation

<p> <b>Data Augmentation  </b>is a technique used to artificially increase dataset size. Take a sample from the dataset, modify it somehow, add it to the original dataset — and now your dataset is one sample larger.

You can do it with all samples in the dataset, and modify each sample several times in a different way — to get a dataset 10, 100, 1000 times larger. You can even create a dataset of infinite size, so your model will never “see” the identical samples during training.

 </p>

```python
image = '/content/drive/MyDrive/Colab Notebooks/cv2_freecodecamp/Resources/Faces/train/Madonna/15.jpg'
```

```python
import tensorflow as tf

from keras.utils import load_img, img_to_array
from keras.preprocessing.image import ImageDataGenerator
```

```python
import matplotlib.pyplot as plt 
```

```python
img = load_img(image , target_size= (200,200))
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30 ,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip =True ,
    width_shift_range = 0.2 ,
    height_shift_range = 0.2 ,
)
```

```python
img = tf.keras.utils.img_to_array(img)
type(img), img.shape
```

```python
input_batch=img.reshape(1,200,200,3)
# 1-> number of image 
```

```python
#  flow  -> It use when we work on single numpy image 
#  flow_from_directory -> To extract all image from certain directory
i=0
for output in datagen.flow(input_batch, batch_size = 1 , save_to_dir = '/content/aug1'):
  i = i+1 
  if i == 10:
    break
```
