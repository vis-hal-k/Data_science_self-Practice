
### Bilateral Bluring
A bilateral filter is used for smoothening images and reducing noise, while preserving edges.

It is very advance and most used bluring method becaue its blur the image by retained edges.

**d**: Diameter of each pixel neighborhood.

**sigmaColor**: Value of \sigma  in the color space. The greater the value, the colors farther to each other will start to get mixed.

**sigmaSpace**: Value of \sigma  in the coordinate space. The greater its value, the more further pixels will mix together, given that their colors lie within the sigmaColor range.

```python
# cv.bilateralFilter(image , d , sigmaColor , sigmaSpace , dst = None , borderType = None)
# d -> diameter | sigmaCOlor ->
bilateral =  cv.bilateralFilter(icat , 20 , 100  , 75)
cv2_imshow(bilateral)
```

## Bitwise Image

```python
import cv2 as cv
import numpy as np
```

```python
blank = np.zeros((400,400) , dtype='uint8' )
rectangle = cv.rectangle(blank.copy() , (30,30) , (370,370) , 255 , -1 )
circle =cv.circle(blank.copy() , (200,200) , 200 , 255 , -1)
#  Bitwise And
bitwise_and = cv.bitwise_and(rectangle , circle)
cv2_imshow(np.hstack((rectangle,circle, bitwise_and)))

# Bitwise Or
bitwiseor = cv.bitwise_or(rectangle , circle)
cv2_imshow(np.hstack((rectangle,circle,bitwiseor)))

# Bitwise XOR  -> not intersecting reason
bitwise_xor = cv.bitwise_xor(rectangle, circle)
cv2_imshow(np.hstack((rectangle,circle,bitwise_xor)))

#  Bitwise NOT
bitwise_not = cv.bitwise_not(rectangle)
cv2_imshow(np.hstack((rectangle,circle,bitwise_not)))
```

##  Masking

 **A mask allows us to focus only on the portions of the image that interests us.**

```python
icat = cv.imread(img)
```

```python
# cv2_imshow(icat)
```

```python
# The Dimension on blank is same as the dimension of image.
blank = np.zeros(icat.shape[:2] , dtype = 'uint8')
cv2_imshow(blank)
# Here we just form the circle in Blank image which name we give is mask.
mask = cv.circle(blank.copy(), (icat.shape[1]//2, icat.shape[0]//2 ) , 100 , 255 ,  -1 )
cv2_imshow(mask)

maskedimage = cv.bitwise_and(icat , icat , mask= mask)
cv2_imshow(maskedimage)
```
