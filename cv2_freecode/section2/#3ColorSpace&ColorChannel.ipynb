```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
i_park = cv.imread(park)
ig_park = cv.cvtColor(i_park , cv.COLOR_BGR2GRAY)
# bgr to hsv
ihsv_park = cv.cvtColor(i_park , cv.COLOR_BGR2HSV)
cv2_imshow(ihsv_park)
# BGR to L*a*b
ilab_park = cv.cvtColor(i_park,cv.COLOR_BGR2LAB)
cv2_imshow(ilab_park)
```

## Color Channels

```python
import cv2 as cv
import numpy as np
img = cv.imread(park)
cv2_imshow(img)
blank = np.zeros(img.shape[:2] , dtype='uint8')
```

#### Split the image

```python
b,g,r = cv.split(img)
blue = cv.merge([b,blank,blank])
green = cv.merge([blank,g,blank])
red = cv.merge([blank,blank,r])
#  It spilt the image in 'blue' , 'green' , 'red' .
# cv2_imshow(b)
# cv2_imshow(g)
# cv2_imshow(r)
cv2_imshow(np.hstack((b,g,r)))
cv2_imshow(np.hstack((blue,green,red)))
```

#### Merge the Image

```python
merged = cv.merge([b,g,r])
# merged = cv.merge([r,g,b]) # This become inverted image
cv2_imshow(merged)
```
