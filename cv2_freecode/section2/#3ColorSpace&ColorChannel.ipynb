```python
park = "/content/drive/MyDrive/Colab Notebooks/cv2_freecodecamp/Resources/Photos/park.jpg"
```

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
i_park = cv.imread(park)
ig_park = cv.cvtColor(i_park , cv.COLOR_BGR2GRAY)
# bgr to hsv
ihsv_park = cv.cvtColor(i_park , cv.COLOR_BGR2HSV)
cv2_imshow(ihsv_park)
# BGR to L*a*b
ilab_park = cv.cvtColor(i_park,cv.COLOR_BGR2LAB)
cv2_imshow(ilab_park)
```
