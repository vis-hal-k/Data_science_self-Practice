# Thresholding

simple thresholding

adaptive thresholding

```python
cat = "/content/drive/MyDrive/Colab Notebooks/cv2_freecodecamp/Resources/Photos/cats.jpg"
```

```python
import cv2 as cv ;
from google.colab.patches import cv2_imshow
import numpy as np ;
import matplotlib.pyplot as plt
```

```python
img = cv.imread(cat);
display(img.shape)
cv2_imshow(img)
gray = cv.cvtColor(img , cv.COLOR_BGR2GRAY)
```

```python
# simple thrsholding
threshold,thresh = cv.threshold(gray,150 ,255,cv.THRESH_BINARY)
cv2_imshow(thresh)
```

```python
# inverse type -> cv.THRESH_BINARY_INV
threshold,thresh_inv = cv.threshold(gray,150 ,255,cv.THRESH_BINARY_INV)
cv2_imshow(thresh_inv)
```

```python
# Adaptive Thresholding
# In this we left the computer to find most optimum value of threshold.
adaptive_thresh = cv.adaptiveThreshold(gray , 255 , cv.ADAPTIVE_THRESH_MEAN_C , cv.THRESH_BINARY , 101 , 2)
#  type 11- > block_size , 3 -> c vale.. 10-> block_size
# cv.ADAPTIVE_THRESH_MEAN_C and cv.ADAPTIVE_THRESH_GAUSSIAN_C
cv2_imshow(adaptive_thresh)
# it really convolve mean.
```

# Edge Detection

Thier is amny to do Edge Detection in this we see :
1. sobe Method
2. Laplacian Method
3. Canny Method

```python
# laplacian
# lap = cv.Laplacian(SRC, ddepth )
lap = cv.Laplacian(gray, cv.CV_64F )
lap = np.uint8(np.absolute(lap))
cv2_imshow(lap)
```

```python
# sobel
xgrad = cv.Sobel(gray,cv.CV_64F , 1, 0)
ygrad = cv.Sobel(gray,cv.CV_64F , 0, 1)
cv2_imshow(xgrad)
print("gap")
cv2_imshow(ygrad)
```

```python
# combine above in soble
comb = cv.bitwise_or(xgrad,ygrad)
cv2_imshow(comb)
```

```python
# Canny
canny = cv.Canny(gray,150,175)
cv2_imshow(canny)
```
